# Splitting Data and PreProcessing

```python
x=heart.drop('target',axis=1)
y=heart['target']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x,y=smote.fit_resample(x,y)
```

```python
x.corrwith(y).sort_values(ascending=False)
```

```python
#dropping columns with little correlation with the target
x.drop(['fbs','chol'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

# Logistic Regression with L1 Regularization

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
lr1=LogisticRegression(penalty='l1',solver='saga',C=1,max_iter=50)
lr1.fit(xtrain,ytrain)
ypred1=lr1.predict(xtest)
print("Logistic Regression with L1 Regularization:")
ypred1tr=lr1.predict(xtrain)
print(f"Training Accuracy:{accuracy_score(ytrain,ypred1tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred1):.3f}")
```

# Logistic Regression With L2 Regularization

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
lr2=LogisticRegression(penalty='l2',solver='lbfgs',C=1,max_iter=1000)
lr2.fit(xtrain,ytrain)
ypred2=lr2.predict(xtest)
print("Logistic Regression with L2 Regularization:")
ypred2tr=lr2.predict(xtrain)
print(f"Training Accuracy:{accuracy_score(ytrain,ypred2tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred2):.3f}")
```

# Logistic Regression With ElasticNets

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
lr3=LogisticRegression(penalty='elasticnet',solver='saga',l1_ratio=0.95,C=1,max_iter=1000)
lr3.fit(xtrain,ytrain)
ypred3=lr3.predict(xtest)
print("Logistic Regression with Elastic Net Regularization:")
ypred3tr=lr3.predict(xtrain)
print(f"Training Accuracy:{accuracy_score(ytrain,ypred3tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred3):.3f}")
```

# Comparision of Penalties

```python
print("Logistic Regression with L1 Regularization:")
print(f"Training Accuracy:{accuracy_score(ytrain,ypred1tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred1):.3f}")
print("\nLogistic Regression with L2 Regularization:")
print(f"Training Accuracy:{accuracy_score(ytrain,ypred2tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred2):.3f}")
print("\nLogistic Regression with Elastic Net Regularization:")
print(f"Training Accuracy:{accuracy_score(ytrain,ypred3tr):.3f}")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred3):.3f}")
```

L1 Regularization and Elastic Net are Achieving Higher Training Accuracy than L2, while all the models are Achieving the same Testing Accuracy, which is lower than the training one so they are all slightly overfit.

# Have you changed other parameter to apply these Penalties

The default solver that is 'lbfgs' only applies on L2 Regularization, it gives error for L1 and Elastic Net, so we have applied a different solver for L1 and Elastic Net that is the 'saga' solver.

# What additional parameter you have changed

Apart from the solver which is not compatible with elastic net and l1, we have also added a new parameter called l1_ratio for the elastic net model, this is necessary because elastic net is a mix between l1 and l2, and l1_ratio is used to control the balance between l1 and l2, so it was necessary to include it

# Briefly discuss the effect of solver on your dataset

The effect of solver is that lbfgs, newton-cg, and newton-cholesky usually give high accuracy and quickly find the best model. The liblinear solver is good for small datasets or when using l1 regularization, but it can become slower with larger datasets or when there are many features. The sag and saga solvers are designed for larger datasets. sag is faster with dense datasets, while saga is more flexible and works well with both dense and sparse datasets. For multiclass problems, only newton-cg, sag, saga and lbfgs handle multinomial loss. Solver is necessary because it used to find the best-fitting model by optimizing the model's parameters

For small to medium-sized datasets, solvers like lbfgs, newton-cg, and newton-cholesky are efficient and perform well, providing fast convergence. For larger datasets, solvers like sag and saga are preferred, with saga being especially versatile as it handles both dense and sparse data. Additionally, when using the multinomial option for multiclass classification, lbfgs, newton-cg, and saga are commonly used as they efficiently handle the complexity of multiple classes.
