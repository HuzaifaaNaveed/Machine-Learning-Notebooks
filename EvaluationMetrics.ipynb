```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title Importing Dataset
diabetes=pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
# @title Checking Null Values
diabetes.isnull().sum()
```

```python
# @title Checking Duplicate Values
diabetes.duplicated().sum()
```

```python
# @title Dropping Duplicates
diabetes.drop_duplicates(inplace=True)
```

```python
# @title Checking Categorical Features
diabetes.head(5)
```

gender and smoking_history are categorical features

```python
diabetes['gender'].unique()
```

```python
diabetes['smoking_history'].unique()
```

```python
# @title Handling Categorical Features
diabetes['gender'] = diabetes['gender'].map({'Male': 1, 'Female': 0, 'Other': 2})
diabetes['smoking_history'] = diabetes['smoking_history'].map({'never': 0, 'No Info': 1, 'current': 2, 'former': 3, 'ever': 4, 'not current': 5})
```

```python
diabetes.head(5)
```

```python
# @title Checking Balance
diabetes['diabetes'].value_counts()
```

since there is a lot of diff so balancing is required

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, balanced_accuracy_score, top_k_accuracy_score, roc_auc_score, roc_curve
```

```python
# @title X Y split
x=diabetes.drop('diabetes', axis=1)
y=diabetes['diabetes']
```

```python
# @title Balancing Data
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
# @title Train Test Split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

since variances are way off, we will need feature scaling

```python
# @title Scaling Data
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
xtrain=pd.DataFrame(xtrain,columns=x.columns)
xtest=pd.DataFrame(xtest,columns=x.columns)
```

```python
# @title Random Forest Evaluation
r=RandomForestClassifier(n_estimators=110,random_state=42)
r.fit(xtrain,ytrain)
ypred1=r.predict(xtest)
print("RANDOM FOREST EVALUATION:")
print(f"Accuracy: {accuracy_score(ytest, ypred1):.3f}")
print(f"Precision: {precision_score(ytest, ypred1):.3f}")
print(f"Recall: {recall_score(ytest, ypred1):.3f}")
print(f"F1 Score: {f1_score(ytest, ypred1):.3f}")
print(f"Balanced Accuracy: {balanced_accuracy_score(ytest, ypred1):.3f}")
```

```python
# @title KNN Evaluation
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
ypred2=knn.predict(xtest)
print("KNN EVALUATION:")
print(f"Accuracy: {accuracy_score(ytest, ypred2):.3f}")
print(f"Precision: {precision_score(ytest, ypred2):.3f}")
print(f"Recall: {recall_score(ytest, ypred2):.3f}")
print(f"F1 Score: {f1_score(ytest, ypred2):.3f}")
print(f"Balanced Accuracy: {balanced_accuracy_score(ytest, ypred2):.3f}")
```

```python
# @title Random Forest Using CV
skf=StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
cva1=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='accuracy')
cvp1=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='precision')
print("RANDOM FOREST EVALUATION USING CV")
print(f"Mean Accuracy:{np.mean(cva1):.3f}")
print(f"Mean Precision:{np.mean(cvp1):.3f}")
```

```python
# @title KNN Using CV
skf=StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
cva1=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='accuracy')
cvp1=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='precision')
print("KNN EVALUATION USING CV")
print(f"Mean Accuracy:{np.mean(cva1):.3f}")
print(f"Mean Precision:{np.mean(cvp1):.3f}")
```

```python
# @title Random Forest ROC Curve Without CV
from sklearn.metrics import auc
prob=r.predict_proba(xtest)[:,1]
fpr,tpr,_=roc_curve(ytest,prob)
area=auc(fpr,tpr)
plt.plot(fpr,tpr,color='green',label=f'Area Under The Curve= {area:.3f}')
plt.plot([0, 1], [0, 1], color='orange', linestyle='--')
plt.title('Random Forest(Without CV) ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
# @title Random Forest ROC Curve With CV
prob=cross_val_predict(r,xtrain,ytrain,cv=skf,method='predict_proba')[:, 1]
fpr,tpr,_=roc_curve(ytrain,prob)
area=auc(fpr,tpr)
plt.plot(fpr,tpr,color='green',label=f'Area Under The Curve= {area:.3f}')
plt.plot([0, 1], [0, 1], color='orange', linestyle='--')
plt.title('Random Forest(With CV) ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

all features are numerical

```python
x=wine.drop('quality',axis=1)
y=wine['quality']
```

```python
corr1=x.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr1, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
cm=x.corr()
hc=set()
for i in range(len(cm.columns)):
    for j in range(i):
        if abs(cm.iloc[i, j]) > 0.5:
            col = cm.columns[i]
            hc.add(col)
x.drop(columns=hc,inplace=True)
```

```python
x.drop('Id',axis=1,inplace=True)
```

```python
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
# @title Train Test Split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

since variances are way off, we will need feature scaling

```python
# @title Scaling Data
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
xtrain=pd.DataFrame(xtrain,columns=x.columns)
xtest=pd.DataFrame(xtest,columns=x.columns)
```

```python
skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
```

```python
# @title Random Forest Evaluation Using CV
r=RandomForestClassifier(n_estimators=110,random_state=42)
r.fit(xtrain,ytrain)
cva=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='accuracy')
cvp=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='precision_weighted')
cvr=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='recall_weighted')
cvf=cross_val_score(r,xtrain,ytrain,cv=skf,scoring='f1_weighted')
print("RANDOM FOREST EVALUATION USING CV")
print(f"Mean Accuracy: {np.mean(cva):.3f}")
print(f"Mean Precision: {np.mean(cvp):.3f}")
print(f"Mean Recall: {np.mean(cvr):.3f}")
print(f"Mean F1 Score: {np.mean(cvf):.3f}")
```

```python
# @title KNN Evaluation Using CV
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)
cva=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='accuracy')
cvp=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='precision_weighted')
cvr=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='recall_weighted')
cvf=cross_val_score(knn,xtrain,ytrain,cv=skf,scoring='f1_weighted')
print("KNN EVALUATION USING CV")
print(f"Mean Accuracy: {np.mean(cva):.3f}")
print(f"Mean Precision: {np.mean(cvp):.3f}")
print(f"Mean Recall: {np.mean(cvr):.3f}")
print(f"Mean F1 Score: {np.mean(cvf):.3f}")
```

# Final Verdict
I checked using both multiclass target and binary class target, and with and without balancing the data. With Balancing the Data, the accuracy, precision, recall and F1-Score were significantly better than leaving it unbalanced, so the better approach is to balance the data. Using Multiclass will lead to better modeling of complex real-world problems, as in most real world cases the target won't be binary. For multiclass target, we will be using weighted precision, recall, and f1-score to evaluate our model. Using weight is essential for multiclass target evaluation. From the models used, Random Forest is giving better results than KNN while we are balancing the data, and using multiclass target, so we can declare that in this case, Random Forest is the Better Model.
