# Applying SVM and KNN

```python
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
xtr, xte, ytr, yte = train_test_split(x, y, test_size=0.1, random_state=42)
svm = SVC(kernel='linear')
knn = KNeighborsClassifier(n_neighbors=3)
ypredsvmtr = cross_val_predict(svm, xtr, ytr, cv=10)
accsvmtr = accuracy_score(ytr, ypredsvmtr)
svm.fit(xtr, ytr)
ypredsvmte = svm.predict(xte)
accsvmte = accuracy_score(yte, ypredsvmte)
print("SVM Training Accuracy: {:.2f}".format(accsvmtr))
print("SVM Test Accuracy: {:.2f}".format(accsvmte))
ypredknntr = cross_val_predict(knn, xtr, ytr, cv=10)
accknntr = accuracy_score(ytr, ypredknntr)
knn.fit(xtr, ytr)
ypredknnte = knn.predict(xte)
accknnte = accuracy_score(yte, ypredknnte)
print("KNN Training Accuracy: {:.2f}".format(accknntr))
print("KNN Test Accuracy: {:.2f}".format(accknnte))
```

# Using Evaluation Metrics

```python
from sklearn.metrics import confusion_matrix, matthews_corrcoef
from prettytable import PrettyTable
import numpy as np
def metrics(yt, yp):
    tn, fp, fn, tp = confusion_matrix(yt, yp).ravel()
    accuracy = (tp + tn) / (tp + tn + fp + fn) * 100
    sensitivity = tp / (tp + fn) * 100
    specificity = tn / (tn + fp) * 100
    geomean = np.sqrt(sensitivity * specificity)
    fdr = fp / (tp + fp) * 100
    for_ = fn / (tn + fn) * 100
    mcc = matthews_corrcoef(yt, yp)
    return accuracy, sensitivity, specificity, geomean, fdr, for_, mcc
accsvmtr, sensvmtr, spesvmtr, gmsvmtr, fdrsvmtr, forsvmtr, mccsvmtr = metrics(ytr, ypredsvmtr)
accsvmte, sensvmte, spesvmte, gmsvmte, fdrsvmte, forsvmte, mccsvmte = metrics(yte, ypredsvmte)
accknntr, senknntr, speknntr, gmknntr, fdrknntr, forknntr, mccknntr = metrics(ytr, ypredknntr)
accknnte, senknnte, speknnte, gmknnte, fdrknnte, forknnte, mccknnte = metrics(yte, ypredknnte)
table = PrettyTable()
table.field_names = ["Parameter", "SVM Tr (%)", "K-NN Tr (%)", "SVM Te (%)", "K-NN Te (%)"]
table.add_row(["Accuracy", f"{accsvmtr:.2f}", f"{accknntr:.2f}", f"{accsvmte:.2f}", f"{accknnte:.2f}"])
table.add_row(["Sensitivity", f"{sensvmtr:.2f}", f"{senknntr:.2f}", f"{sensvmte:.2f}", f"{senknnte:.2f}"])
table.add_row(["Specificity", f"{spesvmtr:.2f}", f"{speknntr:.2f}", f"{spesvmte:.2f}", f"{speknnte:.2f}"])
table.add_row(["Geo Mean", f"{gmsvmtr:.2f}", f"{gmknntr:.2f}", f"{gmsvmte:.2f}", f"{gmknnte:.2f}"])
table.add_row(["FDR", f"{fdrsvmtr:.2f}", f"{fdrknntr:.2f}", f"{fdrsvmte:.2f}", f"{fdrknnte:.2f}"])
table.add_row(["FOR", f"{forsvmtr:.2f}", f"{forknntr:.2f}", f"{forsvmte:.2f}", f"{forknnte:.2f}"])
table.add_row(["MCC", f"{mccsvmtr:.2f}", f"{mccknntr:.2f}", f"{mccsvmte:.2f}", f"{mccknnte:.2f}"])
print("Metrics for Training and Test Sets:")
print(table)
```

# Feature Selection

```python
import pandas as pd
import numpy as np
pos=df[df['diagnosis']==1]
neg=df[df['diagnosis']==0]
fts=df.columns.drop('diagnosis')
fs={}
for f in fts:
    pm = pos[f].mean()
    nm = neg[f].mean()
    om = df[f].mean()
    num = ((pm - om) ** 2) + ((nm - om) ** 2)
    pv = pos[f].var(ddof=1)
    nv = neg[f].var(ddof=1)
    n1 = len(pos)
    n0 = len(neg)
    den = (1 / (n1 - 1)) * pv + (1 / (n0 - 1)) * nv
    fsc = num / den if den != 0 else 0
    fs[f] = fsc
fs=pd.DataFrame(list(fs.items()),columns=['Feature','F']).sort_values(by='F', ascending=False)
print(fs)
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
def calc_metrics(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sen = tp / (tp + fn) * 100
    spe = tn / (tn + fp) * 100
    ppv = tp / (tp + fp) * 100
    npv = tn / (tn + fn) * 100
    return sen, spe, ppv, npv
met = {'Model': [], '50-50': [], '70-30': [], '80-20': []}
def calc_met(feat, X, y, ts):
    X_tr, X_ts, y_tr, y_ts = train_test_split(X[feat], y, test_size=ts, random_state=42)
    svm = SVC(kernel='rbf', random_state=42)
    svm.fit(X_tr, y_tr)
    y_pred = svm.predict(X_ts)
    sen, spe, ppv, npv = calc_metrics(y_ts, y_pred)
    return sen, spe, ppv, npv
feat=mdl_df.loc[4,'Features'].split(', ')
sen_5050, spe_5050, ppv_5050, npv_5050 = calc_met(feat, x, y, ts=0.5)
sen_7030, spe_7030, ppv_7030, npv_7030 = calc_met(feat, x, y, ts=0.3)
sen_8020, spe_8020, ppv_8020, npv_8020 = calc_met(feat, x, y, ts=0.2)
met['Model'].append(5)
met['50-50'].append(f"Sen: {sen_5050:.2f}, Spec: {spe_5050:.2f}, PPV: {ppv_5050:.2f}, NPV: {npv_5050:.2f}")
met['70-30'].append(f"Sen: {sen_7030:.2f}, Spec: {spe_7030:.2f}, PPV: {ppv_7030:.2f}, NPV: {npv_7030:.2f}")
met['80-20'].append(f"Sen: {sen_8020:.2f}, Spec: {spe_8020:.2f}, PPV: {ppv_8020:.2f}, NPV: {npv_8020:.2f}")
met_df=pd.DataFrame(met)
print(met_df)
```

# SVM with All Kernels

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
xtr, xte, ytr, yte = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xte=sc.transform(xte)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for k in kernels:
    svm=SVC(kernel=k)
    svm.fit(xtr, ytr)
    ytr_pred=svm.predict(xtr)
    yte_pred=svm.predict(xte)
    tr_acc=accuracy_score(ytr, ytr_pred)
    te_acc=accuracy_score(yte, yte_pred)
    print("Kernel:",k)
    print(f"Train Accuracy: {tr_acc:.4f}, Test Accuracy: {te_acc:.4f}")
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
c_values = [0.1, 1, 10]
gamma_values = [0.1, 1, 10]
for k in kernels:
  for c in c_values:
    for gamma in gamma_values:
      svm = SVC(kernel=k, C=c, gamma=gamma)
      svm.fit(xtr, ytr)
      ytr_pred = svm.predict(xtr)
      yte_pred = svm.predict(xte)
      tr_acc = accuracy_score(ytr, ytr_pred)
      te_acc = accuracy_score(yte, yte_pred)
      print(f"Kernel: {k}, C: {c}, gamma: {gamma}")
      print(f"Train Accuracy: {tr_acc:.4f}, Test Accuracy: {te_acc:.4f}")
      print()
```
