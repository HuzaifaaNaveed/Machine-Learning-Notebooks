```python
from sklearn.preprocessing import StandardScaler
x=df.drop('target',axis=1)
s=StandardScaler()
x=s.fit_transform(x)
x=pd.DataFrame(x)
x
```

```python
# @title Train Test Split
from sklearn.model_selection import train_test_split
y=df['target']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrainval,xtestval,ytrainval,ytestval=train_test_split(xtrain,ytrain,test_size=0.3,random_state=0)
```

```python
acctr=[]
accts=[]
```

```python
# @title Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
r=RandomForestClassifier(random_state=0)
r.fit(xtrain, ytrain)
prte=r.predict(xtest)
prtr=r.predict(xtrain)
acctr.append(accuracy_score(ytrain,prtr))
accts.append(accuracy_score(ytest,prte))
print("Random Forest:")
print(f"Testing Accuracy: {accuracy_score(ytest,prte):.3f}")
print(f"Training Accuracy: {accuracy_score(ytrain,prtr):.3f}")
```

```python
# @title AdaBoost
from sklearn.ensemble import AdaBoostClassifier
a=AdaBoostClassifier(algorithm='SAMME',random_state=0)
a.fit(xtrain,ytrain)
prte=a.predict(xtest)
prtr=a.predict(xtrain)
acctr.append(accuracy_score(ytrain,prtr))
accts.append(accuracy_score(ytest,prte))
print("AdaBoost:")
print(f"Testing Accuracy: {accuracy_score(ytest, prte):.3f}")
print(f"Training Accuracy: {accuracy_score(ytrain, prtr):.3f}")
```

```python
# @title XGBoost
from xgboost import XGBClassifier
x=XGBClassifier(random_state=0)
x.fit(xtrain, ytrain)
prte=x.predict(xtest)
prtr=x.predict(xtrain)
acctr.append(accuracy_score(ytrain,prtr))
accts.append(accuracy_score(ytest,prte))
print("XGBoost:")
print(f"Testing Accuracy: {accuracy_score(ytest, prte):.3f}")
print(f"Training Accuracy: {accuracy_score(ytrain, prtr):.3f}")
```

```python
# @title Comparing Testing and Training Accuracies
import numpy as np
import matplotlib.pyplot as plt
models=['Random Forest', 'AdaBoost', 'XGBoost']
x1=np.arange(len(models))
width=0.35
plt.bar(x1-width/2,acctr, width, label='Training Accuracy', color='lightblue')
plt.bar(x1+width/2,accts, width, label='Testing Accuracy', color='salmon')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Training and Testing Accuracy Comparison')
plt.xticks(x1, models)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# @title Using Same Dataset as Q1
df=pd.read_csv('heart.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# @title Extracting restecg And oldpeak
x=df[['restecg','oldpeak']]
y=df['target']
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x=s.fit_transform(x)
x=pd.DataFrame(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# @title Training a Voting Classifier using Decision Tree, KNN, Random Forest and XGboost
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
d=DecisionTreeClassifier(random_state=42)
k=KNeighborsClassifier(n_neighbors=7)
r=RandomForestClassifier(random_state=42,n_estimators=110)
xg=XGBClassifier(random_state=42)
hard=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='hard')
soft=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='soft')
hard.fit(xtrain,ytrain)
soft.fit(xtrain,ytrain)
ypredhard=hard.predict(xtest)
ypredsoft=soft.predict(xtest)
tr1=hard.predict(xtrain)
tr2=soft.predict(xtrain)
print("Hard Voting: ")
print(f"Training Accuracy: {accuracy_score(ytrain, tr1):.3f}")
print(f"Testing Accuracy: {accuracy_score(ytest,ypredhard):.3f}")
print("\nSoft Voting: ")
print(f"Training Accuracy: {accuracy_score(ytrain, tr2):.3f}")
print(f"Testing Accuracy: {accuracy_score(ytest,ypredsoft):.3f}")
```

```python
# @title Checking which Voting Parameter is Giving Best Accuracy
if accuracy_score(ytest,ypredhard)>accuracy_score(ytest,ypredsoft):
  print("Hard Voting Gives Better Accuracy")
  print(f"Hard Voting Accuracy: {accuracy_score(ytest,ypredhard):.3f}")
else:
  print("Soft Voting Gives Better Accuracy")
  print(f"Soft Voting Accuracy: {accuracy_score(ytest,ypredsoft):.3f}")
```

I will be checking the weights manually to determine the best results

```python
# @title Checking the best weights for Hard Voting
# checking best weights manually
weightedvote=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='hard',weights=[1,4,1,1])
weightedvote.fit(xtrain,ytrain)
ypred3=weightedvote.predict(xtest)
trh=hard.predict(xtrain)
print("Hard Voting Accuracy for Best Weights:")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred3):.3f}")
print(f"Training Accuracy:{accuracy_score(ytrain,trh):.3f}")
```

Hard Voting Accuracy increases when we increase the weight of knn, so we will assign weight of 4 to knn. After using many possible weights, using 1,4,1,1 gave the best accuracy.

```python
# @title Checking the best weights for Soft Voting
# checking best weights manually
weightedvote=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='soft',weights=[1,1,1,1])
weightedvote.fit(xtrain,ytrain)
ypred2=weightedvote.predict(xtest)
trs=hard.predict(xtrain)
print("Soft Voting Accuracy for Best Weights:")
print(f"Testing Accuracy:{accuracy_score(ytest,ypred2):.3f}")
print(f"Training Accuracy:{accuracy_score(ytrain,trs):.3f}")
```

Soft weighting accuracy remains same for the different no. of weight combinations I tried, so we will keep all the weights the same i.e 1,1,1,1

```python
# @title Soft Voting Bias Variance Tradeoff
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
soft=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='soft',weights=[1,1,1,1])
tre=[]
tse=[]
for n in range(1,10):
    soft.set_params(knn__n_neighbors=n)
    soft.fit(xtrain,ytrain)
    ypredtr=soft.predict(xtrain)
    ypredts=soft.predict(xtest)
    tre.append(mean_squared_error(ytrain,ypredtr))
    tse.append(mean_squared_error(ytest,ypredts))
plt.plot(range(1,10),tre,label='Training Error',marker='o',color='blue')
plt.plot(range(1,10),tse,label='Testing Error',marker='o',color='red')
plt.xlabel('Model Complexity (KNN No. of Neighbours)')
plt.ylabel('Error')
plt.legend()
plt.title('Soft Voting Bias-Variance Tradeoff')
plt.show()
```

```python
# @title Hard Voting Bias Variance Tradeoff
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
soft=VotingClassifier(estimators=[('dt',d),('knn',k),('rf',r),('xgb',xg)],voting='hard',weights=[1,4,1,1])
tre=[]
tse=[]
for n in range(1,10):
    soft.set_params(knn__n_neighbors=n)
    soft.fit(xtrain,ytrain)
    ypredtr=soft.predict(xtrain)
    ypredts=soft.predict(xtest)
    tre.append(mean_squared_error(ytrain,ypredtr))
    tse.append(mean_squared_error(ytest,ypredts))
plt.plot(range(1,10),tre,label='Training Error',marker='o',color='blue')
plt.plot(range(1,10),tse,label='Testing Error',marker='o',color='red')
plt.xlabel('Model Complexity (KNN No. of Neighbours)')
plt.ylabel('Error')
plt.legend()
plt.title('Hard Voting Bias-Variance Tradeoff')
plt.show()
```

```python
# @title Voting Classifier for Random Forest and AdaBoost
r=RandomForestClassifier(n_estimators=110,random_state=0)
a=AdaBoostClassifier(algorithm='SAMME',random_state=0)
hard=VotingClassifier(estimators=[('rf',r),('ada',a)],voting='hard',weights=[3,1])
soft=VotingClassifier(estimators=[('rf',r),('ada',a)],voting='soft',weights=[3,1])
hard.fit(xtrain,ytrain)
soft.fit(xtrain,ytrain)
ypredhard=hard.predict(xtest)
ypredsoft=soft.predict(xtest)
tr1=hard.predict(xtrain)
tr2=soft.predict(xtrain)
print("Hard Voting: ")
print(f"Training Accuracy: {accuracy_score(ytrain, tr1):.3f}")
print(f"Testing Accuracy: {accuracy_score(ytest,ypredhard):.3f}")
print("\nSoft Voting: ")
print(f"Training Accuracy: {accuracy_score(ytrain, tr2):.3f}")
print(f"Testing Accuracy: {accuracy_score(ytest,ypredsoft):.3f}")
```

```python
# @title Individual Vs Ensemble Graphs
r.fit(xtrain,ytrain)
rtr=accuracy_score(ytrain,r.predict(xtrain))
rts=accuracy_score(ytest,r.predict(xtest))
xg.fit(xtrain,ytrain)
xtr=accuracy_score(ytrain,xg.predict(xtrain))
xts=accuracy_score(ytest,xg.predict(xtest))
hard.fit(xtrain,ytrain)
etr=accuracy_score(ytrain,hard.predict(xtrain))
ets=accuracy_score(ytest,hard.predict(xtest))
soft.fit(xtrain,ytrain)
str1=accuracy_score(ytrain,soft.predict(xtrain))
sts=accuracy_score(ytest,soft.predict(xtest))
a.fit(xtrain,ytrain)
atr=accuracy_score(ytrain,a.predict(xtrain))
ats=accuracy_score(ytest,a.predict(xtest))
acctr=[rtr,xtr,atr,etr,str1]
accts=[rts,xts,ats,ets,sts]
models=['Random Forest', 'XGBoost','AdaBoost','Hard Voting', 'Soft Voting']
x1=np.arange(len(models))
width=0.35
plt.bar(x1-width/2,acctr, width, label='Training Accuracy', color='lightblue')
plt.bar(x1+width/2,accts, width, label='Testing Accuracy', color='salmon')
plt.plot(x1 + width/2, accts,color='grey',linestyle='-',linewidth=1)
plt.plot(x1 - width/2, acctr,color='grey',linestyle='-',linewidth=1)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Training and Testing Accuracy Comparison')
plt.xticks(x1, models)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
