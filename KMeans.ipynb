```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
paris=scaler.fit_transform(paris)
```

```python
from sklearn.cluster import KMeans
ine=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(paris)
    ine.append(kmeans.inertia_)
plt.plot(range(1, 11),ine,marker='o',color='salmon')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

# Optimal K=3

```python
kmeans=KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(paris)
```

```python
pred=kmeans.predict(paris)
parisClusters=pd.read_csv('ParisHousing.csv')
parisClusters.drop('Unnamed: 0',axis=1,inplace=True)
parisClusters['Cluster']=pred
parisClusters
```

```python
shopping_data=pd.read_csv('Online Retail.csv')
```

```python
shopping_data
```

```python
shopping_data.describe()
```

```python
shopping_data.isnull().sum()
```

```python
shopping_data.duplicated().sum()
```

```python
shopping_data.drop_duplicates(inplace=True)
```

```python
shopping_data=shopping_data.dropna(subset=["CustomerID"])
```

```python
shopping_data.isnull().sum()
```

```python
shopping_data=shopping_data[shopping_data['UnitPrice']>0]
```

```python
shopping_data=shopping_data[shopping_data['Quantity']>0]
```

```python
shopping_data.drop(['Description','InvoiceDate','InvoiceNo'],axis=1,inplace=True)
```

```python
shopping_data.drop('StockCode',axis=1,inplace=True)
```

```python
shopping_data.drop('Country',axis=1,inplace=True)
```

```python
shopping_data["TotalAmount"]=shopping_data["Quantity"]*shopping_data["UnitPrice"]
shop=shopping_data.groupby('CustomerID').agg(totalItems=('Quantity','sum'),totalSpent=('TotalAmount','sum'))
shop=shop.reset_index()
shop
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
shop2=scaler.fit_transform(shop)
```

```python
from sklearn.cluster import KMeans
ine=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(shop2)
    ine.append(kmeans.inertia_)
plt.plot(range(1, 11),ine,marker='o',color='salmon')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

# Optimal Clusters=4

```python
kmeans=KMeans(n_clusters=4,init='k-means++', random_state=42)
kmeans.fit(shop2)
```

```python
pred=kmeans.predict(shop2)
shop['Cluster']=pred
shop
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.scatterplot(x='totalSpent', y='totalItems', hue='Cluster', data=shop, palette='magma', s=100)
plt.title('Customer Segmentation based on Spending and Items Purchased')
plt.xlabel('Total Spent')
plt.ylabel('Total Items')
plt.show()
```

```python
shop.groupby('Cluster').agg(totalSpent=('totalSpent','sum'),totalItems=('totalItems','sum'))
```

```python
shop.groupby('Cluster').agg(totalSpent=('totalSpent','mean'),totalItems=('totalItems','mean'))
```

```python
shop.groupby('Cluster').agg(totalSpent=('totalSpent','median'),totalItems=('totalItems','median'))
```
